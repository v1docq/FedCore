# FedCore example: запуск модуля экспорта через Docker

Этот ноутбук описывает и проверяет структуру demo-сервиса экспорта моделей через Docker.

Пример включает FastAPI-приложение `app.py`, `Dockerfile`, `docker-compose.yml`, endpoint `/health`, endpoint `/upload`, endpoint `/export` и endpoint `/files`.

> В demo-версии `/export` имитирует экспорт: копирует файл и меняет расширение. В промышленной версии этот endpoint может вызывать FedCore pipeline экспорта модели.

## 1. Ожидаемая структура файлов

```python
from pathlib import Path

project_dir = Path("deployment/model_exporter")

expected_files = [
    project_dir / "app.py",
    project_dir / "requirements.txt",
    project_dir / "Dockerfile",
    project_dir / "docker-compose.yml",
    project_dir / "README.md",
]

for path in expected_files:
    print(path)
```

## 2. Запуск сервиса

```python
print("cd deployment/model_exporter")
print("docker compose up --build")
```

## 3. Проверка состояния сервиса

```python
print("curl http://localhost:5000/health")
```

## 4. Загрузка модельного файла

```python
print('curl -X POST "http://localhost:5000/upload" -F "file=@demo_model.pt"')
```

## 5. Запуск demo-экспорта

```python
print('curl -X POST "http://localhost:5000/export?filename=demo_model.pt&target_format=onnx"')
```

## Выводы по демонстрационному примеру

В примере подготовлен минимальный Docker-сервис для демонстрации модуля экспорта моделей.

Сервис закрывает пользовательский сценарий:

- запуск контейнера через `docker compose up --build`;
- проверка состояния через `/health`;
- загрузка модельного файла через `/upload`;
- запуск demo-экспорта через `/export`;
- просмотр файлов через `/files`.

Такой пример можно использовать как базовую точку входа для дальнейшего подключения реального FedCore pipeline экспорта моделей в ONNX, TensorRT, TFLite / LiteRT, OpenVINO или другие форматы.
